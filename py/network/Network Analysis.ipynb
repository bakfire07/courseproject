```python
import networkx as nx
import visJS2jupyter.visJS_module as v
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline  
# read the file
```

```python
# read the file
fs = open("/Users/balamurali/Downloads/TMSMA-Network - Form responses.tsv",'r')
G = nx.DiGraph()
fs.readline()
edges = list()
nodes = list()
for line in fs:
    line_info = line.split('\t')
    primary_node = line_info[1]
    secondary_node_arr = line_info[2].split(',')
    elist = [(primary_node.replace('@iimv.ac.in',''),secondary_node.strip().replace('@iimv.ac.in','')) for secondary_node in secondary_node_arr if len(secondary_node.strip()) >0 ]
    nlist = [secondary_node.strip().replace('@iimv.ac.in','') for secondary_node in secondary_node_arr if len(secondary_node.strip()) >0 ]
    edges +=  elist
    nodes += nlist
nodes = list(set(nodes))
G.add_nodes_from(nodes)
G.add_edges_from(edges)
    
# get the nodes and edges
nodes = G.nodes()
edges = G.edges()
```

```python

# visualize
pos=nx.spring_layout(G,k=0.1,iterations=50,scale=10)
nx.draw(G, with_labels = True)
plt.show()
nx.write_gexf(G, "/Users/balamurali/Downloads/tmsma.gexf")
```

```python
# find the popular fellow in the class
from operator import itemgetter
popularity_list = sorted(G.degree_iter(),key=itemgetter(1),reverse=True)
print '#################################\n\n'
print popularity_list[0]
#print 'most popular person: {}'.format(str(popularity_list[0]) 
print '\n\n#################################'
```

```python
# find the shortest path between two people
import random
node1 = random.choice(G.nodes())
node2 = random.choice(G.nodes())
print '#################################\n\n'
print 'Finding the shortest path between {} and {}'.format(node1,node2)
print '\n\n#################################'
print nx.shortest_path(G,source=node1,target=node2)

```

```python
# strongly connected components
strongly_connected_components = nx.strongly_connected_components(G)
print 'Strongly connected groups are:'
for a in strongly_connected_components:
    if len(a)>1:
        print a
```

```python
# transitivity
transitivity = nx.transitivity(G)

print transitivity
    
```
